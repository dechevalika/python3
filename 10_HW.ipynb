```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.head(10)
```

```python
%matplotlib inline
sns.heatmap(penguins.corr())
```

```python
# отношение длины плавника к массе тела по полу, соответственно особи муж пола крупнее и с длинными плавниками
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex")
```

```python
# отношение длины клюва к ширине клюва по разновидностям, соответственно на графике видно, что у Адели короткие клювы и широкие, у Чинстрап длинные и широкие и у Дженту длинные и узкие
sns.scatterplot(data=penguins, x="bill_depth_mm", y="bill_length_mm", style="species")
```

```python
# отношение массы тела к виду, на графике видно, что Чинстрап и Адели в одной весовой категории, но Чинстрап находятся только на острове Dream, когда как Адели на всех островах, а Дженту по массе больше, но находятся на острове Biscoe
sns.scatterplot(data=penguins, x="body_mass_g", y="species", size="island")
```

```python
sns.histplot(data=penguins, x="body_mass_g")
```

```python
sns.histplot(penguins, x="island", hue="species")
```

```python
cols = ['bill_length_mm', 'bill_depth_mm']
g = sns.PairGrid(penguins[cols])
g.map(sns.scatterplot)
```
